```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
import glob

from natsort import natsorted
```

```python
names = [];
for name in glob.glob('processor0/*/uniform/functionObjects/functionObjectProperties'):
    names.append(name)
len(names)
```

```python
def readFile(file):
    with open(file) as f:
        lines = f.readlines()
    line = lines[24]
    return float(line[28:-3])
```

```python
Cd = []
t = []
for name in natsorted(names):
    Cd.append(readFile(name))
    t.append(float(name.split(os.sep)[1]))
Cd;
```

```python
plt.plot(t[3:],Cd[3:],'bo')
plt.plot(t[3:],Cd[3:],'g',linewidth=3)
plt.xticks(np.arange(0,max(t), step=max(t)/10));
plt.title('Drag coefficient during the simulation')
plt.xlabel('t [s]')
plt.ylabel('Cd [ ]')
plt.grid(True)
plt.savefig('DragCoefficientGraph.png')
```
